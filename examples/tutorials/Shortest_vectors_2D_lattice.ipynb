# Shortest vector in 2D lattices

```julia
using LatticeAlgorithms
using LinearAlgebra
using Plots

function shortest_vector_parallelogram(λ, δ, θ, name; v1 = [1 0], xlim = 10, ylim = 10)
    """Find the shortest vector in a parallelogram
    
    Args:
        λ: lattice constant
        δ: ratio between the length of basis vector v2 and v1 
        θ: angle between the two basis vectors
        v1 = [1 0]: The basis vector v1    
    
    Note: 
        If there are multiple shortest vectors with the same length, we only return one.
    
    """
    
    v2 = δ * [cos(θ) sin(θ)] # basis vector v2
    M = λ * [v1 ; v2] # generator matrix 
    
#     u3_hat = Decode_ShortestVector(M)
    
#     sv = transpose(u3_hat) * M
    
    sv = shortest_vector(M)
    
#     print(all_closest_points)
    
    pts = get_grid_points_2D_lattice(M, xlim, ylim)
    
#     println(sv)
        
    g = scatter([pt[1] for pt in pts], [pt[2] for pt in pts], markercolor=:red, aspect_ratio=:equal, label="")
    scatter!([sv[1]], [sv[2]], marker = :square, markercolor = :black, label="Shortest Vector from the origin")
    scatter!([0], [0], marker = :circle, markercolor = :blue, label="Origin")    
    xlims!((-xlim, xlim))
    ylims!((-ylim, ylim))

    return g    
    
end
```

## Square lattice

```julia
shortest_vector_parallelogram(1, 1, π/2, "square", xlim=3, ylim=3)
```

## Rectangular lattice

```julia
shortest_vector_parallelogram(1, 2, π/2, "rectangular", xlim=3, ylim=3)
```

## Triangular lattice

```julia
shortest_vector_parallelogram(1, 1, π/3, "triangular", xlim=3, ylim=3)
```

## General parallelogram

```julia
shortest_vector_parallelogram(1, 0.5, π/3, "general_parallelogram_1", xlim=2, ylim=2)
```
