# All closest points for 2D lattices

In this notebook, we demonstrate how to find all the closest points for a given ${\bf x}\in\mathbb{R}^n$ for several 2D lattices.

Finding all the closest points is a generalization of the problem of finding a closest point for a given ${\bf x}\in\mathbb{R}^n$. For the latter problem, we output the solution immediately after we find one, while for the former, we continue to find the next possible closest point for ${\bf x}$, until none can be found further. Finding all the closest point is essential to find the relevant vectors of the lattice, as explained in [Agrell et. al.](https://publications.lib.chalmers.se/records/fulltext/14990/local_14990.pdf).

```julia
using LatticeAlgorithms
using LinearAlgebra
using Plots

function all_closest_points_parallelogram(x, λ, δ, θ, name; v1 = [1 0], xlim = 10, ylim = 10)
    """Find all the cloest lattice points for a given x in a parallelogram
    
    Args:
        x: The given vector
        λ: lattice constant
        δ: ratio between the length of basis vector v2 and v1 
        θ: angle between the two basis vectors
        v1 = [1 0]: The basis vector v1    
    
    """
    
    v2 = δ * [cos(θ) sin(θ)] # basis vector v2
    M = λ * [v1 ; v2] # generator matrix 
    
#     U3_hat = all_closest_points(x, M)
    
    all_cpts = all_closest_points(vec(x), M) # [u3 * M for u3 in U3_hat]
    
#     print(all_closest_points)
    
    pts = get_grid_points_2D_lattice(M, xlim, ylim)
        
    g = scatter([pt[1] for pt in pts], [pt[2] for pt in pts], markercolor=:red, aspect_ratio=:equal, label="")
    scatter!([x[1]], [x[2]], markercolor=:blue, label="Target point x")
    scatter!([pt[1] for pt in all_cpts], [pt[2] for pt in all_cpts], marker = :square, markercolor = :black, label="All closest points")
    xlims!((-xlim, xlim))
    ylims!((-ylim, ylim))

    return g    
    
end
```

## Square lattice

```julia
all_closest_points_parallelogram([1 2], 1, 1, π/2, "square_1", xlim=3, ylim=3)
```

```julia
all_closest_points_parallelogram([0 1/2], 1, 1, π/2, "square_2", xlim=3, ylim=3)
```

```julia
all_closest_points_parallelogram([0 1/4], 1, 1, π/2, "square_3", xlim=3, ylim=3)
```
