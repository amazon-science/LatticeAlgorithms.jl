```julia
old_num_samples = data_surf_sq_33_33["num_samples"]
old_t_list = data_surf_sq_33_33["t_list"]
old_c_list = data_surf_sq_33_33["c_list"]
old_p_list = data_surf_sq_33_33["p_list"];
[length(old_t_list), length(old_p_list), length(old_c_list)]
```

```julia
# data_new_33 = load("raw_data/33_data/bsv_surface_square_33_33_0.591_0.6_1012_1161811862167834384.jld2")

c_list = Dict()
p_list = Dict()
t_list = Dict()

total_num_samples = 0
folder = "raw_data/37_39_data/"
for file in readdir(folder)
    if file[end-3:end] == "jld2"
        newdata = load("$folder/$file")
        newdrange = newdata["drange"]
        newσrange = newdata["σrange"]
#         @assert drange == newdrange
#         @assert σrange == newσrange
        new_c_list = newdata["c_list"]
        new_p_list = newdata["p_list"]
        new_t_list = newdata["t_list"]
        
        new_num_samples = newdata["num_samples"]
        
        map!(v->v.*new_num_samples, values(new_p_list))
        map!(v->v.*new_num_samples, values(new_c_list))
        map!(v->v.*new_num_samples, values(new_t_list))
                
        total_num_samples += new_num_samples
        
#         println(new_p_list)
        
        p_list = merge(+, p_list, new_p_list)
        c_list = merge(+, c_list, new_c_list)
        t_list = merge(+, t_list, new_t_list)
    end
end
# println(p_list)

total_num_samples = total_num_samples/2

map!(v->v./total_num_samples, values(c_list))
map!(v->v./total_num_samples, values(p_list))
map!(v->v./total_num_samples, values(t_list))
# println(p_list)
```

```julia
[length(t_list), length(p_list), length(c_list)]
```

```julia
display([c_list[k] - old_c_list[k] for (k, v) in c_list])
println()
display([p_list[k] - old_p_list[k] for (k, v) in p_list])
println()
display([t_list[k] - old_t_list[k] for (k, v) in t_list])
```

```julia
new_num_samples = old_num_samples + total_num_samples
```

```julia
for (k, v) in old_c_list
    if k ∉ keys(c_list)
        c_list[k] = old_c_list[k]
        t_list[k] = old_t_list[k]
        p_list[k] = old_p_list[k]
    end
    c_list[k] = (c_list[k] * total_num_samples + old_c_list[k] * old_num_samples) / new_num_samples
    p_list[k] = (p_list[k] * total_num_samples + old_p_list[k] * old_num_samples) / new_num_samples
    t_list[k] = (t_list[k] * total_num_samples + old_t_list[k] * old_num_samples) / new_num_samples
end
```

```julia
σrange_surf_sq = vcat(0.5:0.01:0.59, 0.591:0.001:0.607, exp(-1/2))

c_list_data_surf_sq = merge(
    data_surf_sq_3_21["c_list"],
    data_surf_sq_23_23["c_list"],
    data_surf_sq_25_25["c_list"],
    data_surf_sq_27_27["c_list"],
    data_surf_sq_29_29["c_list"],
    data_surf_sq_31_31["c_list"],
    data_surf_sq_33_33["c_list"],
    data_surf_sq_35_35["c_list"],
    data_surf_sq_37_39["c_list"],
    data_surf_sq_3_39_threshold["c_list"],
    data_surf_sq_3_23_596["c_list"],
)

p_list_data_surf_sq = merge(
    data_surf_sq_3_21["p_list"],
    data_surf_sq_23_23["p_list"],
    data_surf_sq_25_25["p_list"],
    data_surf_sq_27_27["p_list"],
    data_surf_sq_29_29["p_list"],
    data_surf_sq_31_31["p_list"],
    data_surf_sq_33_33["p_list"],
    data_surf_sq_35_35["p_list"],
    data_surf_sq_37_39["p_list"],
    data_surf_sq_3_39_threshold["p_list"],
    data_surf_sq_3_23_596["p_list"],
)

t_list_data_surf_sq = merge(
    data_surf_sq_3_21["t_list"],
    data_surf_sq_23_23["t_list"],
    data_surf_sq_25_25["t_list"],
    data_surf_sq_27_27["t_list"],
    data_surf_sq_29_29["t_list"],
    data_surf_sq_31_31["t_list"],
    data_surf_sq_33_33["t_list"],
    data_surf_sq_35_35["t_list"],
    data_surf_sq_37_39["t_list"],
    data_surf_sq_3_39_threshold["t_list"],
    data_surf_sq_3_23_596["t_list"],
)

drange_surf_sq = 1 : 2 : 39
for σ in σrange_surf_sq
    for d in drange_surf_sq
        if d == 1 && σ ∈ collect(0.5:0.01:0.6)
            c_list_data_surf_sq[[σ, d]] = data_one_mode_square_1["c_list"][σ]
            t_list_data_surf_sq[[σ, d]] = data_one_mode_square_1["t_list"][σ]
            p_list_data_surf_sq[[σ, d]] = sqrt(data_one_mode_square_1["p_list"][σ][1])
        elseif d == 1 && σ ∈ collect(0.601:0.001:0.607)
            c_list_data_surf_sq[[σ, d]] = data_one_mode_square_2["c_list"][σ]
            t_list_data_surf_sq[[σ, d]] = data_one_mode_square_2["t_list"][σ]
            p_list_data_surf_sq[[σ, d]] = sqrt(data_one_mode_square_2["p_list"][σ][1])
        else
            if [σ, d] ∉ keys(c_list_data_surf_sq)
                c_list_data_surf_sq[[σ, d]] = NaN
                t_list_data_surf_sq[[σ, d]] = NaN
                p_list_data_surf_sq[[σ, d]] = NaN
            end
        end
    end
end
```
