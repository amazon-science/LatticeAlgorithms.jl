```julia
using LatticeAlgorithms

n = 10 # Dimension of the lattice
M = rand(n, n) # A lattice generated by a random matrix
lll_basis = lll(M)
@assert lll_basis.T * lll_basis.L * lll_basis.Q ≈ M

# True
```

```julia
islllreduced(lll_basis.L)
```

```julia
kz_basis = kz(M)
@assert kz_basis.T * kz_basis.L * kz_basis.Q ≈ M # true
```

```julia
using LatticeAlgorithms
n = 20 # Dimension of the lattice

M = Dn(n)
x = rand(n)

@time y1 = closest_point(x, M)
@time y2 = closest_point_Dn(x)
@assert y1 ≈ y2
```
